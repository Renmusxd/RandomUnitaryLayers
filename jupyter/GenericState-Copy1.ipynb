```python
import numpy
import py_entropy
from matplotlib import pyplot
```

```python
sizes = []
no_pre_time = []
pre_time = []

Ls = list(range(5,12))
for i in Ls:
    g = py_entropy.GenericMultiDefectState(i, i, [1,2,1], parallel_matrix_mul=False, num_experiments=16)
    sizes.append(g.get_precompute_size())
    ret = %timeit -o g.apply_layer()
    no_pre_time.append(ret)

    g = py_entropy.GenericMultiDefectState(i, i, [1,2,1], parallel_matrix_mul=False, num_experiments=16)
    g.precompute_connections()
    ret = %timeit -o g.apply_layer()
    pre_time.append(ret)
```

```python
pyplot.plot(Ls, [x.average for x in no_pre_time])
pyplot.plot(Ls, [x.average for x in pre_time])
pyplot.yscale('log')
pyplot.xlabel('L')
pyplot.ylabel('Time to Run')
pyplot.show()

pyplot.plot(Ls, sizes)
pyplot.yscale('log')
pyplot.xlabel('L')
pyplot.ylabel('Precompute Memory Cost')
pyplot.show()
```

```python
sizes = []
no_pre_time = []
pre_time = []

Ls = list(range(5,14))
for i in Ls:
    g = py_entropy.GenericMultiDefectState(i, i//2, [1,2], parallel_matrix_mul=False, num_experiments=16)
    sizes.append(g.get_precompute_size())
    ret = %timeit -o g.apply_layer()
    no_pre_time.append(ret)

    g = py_entropy.GenericMultiDefectState(i, i//2, [1,2], parallel_matrix_mul=False, num_experiments=16)
    g.precompute_connections()
    ret = %timeit -o g.apply_layer()
    pre_time.append(ret)
    
par_sizes = []
par_no_pre_time = []
par_pre_time = []

for i in Ls:
    g = py_entropy.GenericMultiDefectState(i, i//2, [1,2], parallel_matrix_mul=True, num_experiments=16)
    par_sizes.append(g.get_precompute_size())
    ret = %timeit -o g.apply_layer()
    par_no_pre_time.append(ret)

    g = py_entropy.GenericMultiDefectState(i, i//2, [1,2], parallel_matrix_mul=True, num_experiments=16)
    g.precompute_connections()
    ret = %timeit -o g.apply_layer()
    par_pre_time.append(ret)
```

```python
pyplot.plot(Ls, [x.average for x in no_pre_time], label='Std')
pyplot.plot(Ls, [x.average for x in pre_time], label='Pre')
pyplot.plot(Ls, [x.average for x in par_no_pre_time], label='Par Std')
pyplot.plot(Ls, [x.average for x in par_pre_time], label='Par Pre')
pyplot.yscale('log')
pyplot.xlabel('L')
pyplot.ylabel('Time to Run')
pyplot.legend()
pyplot.show()


pyplot.plot(Ls, sizes)
pyplot.yscale('log')
pyplot.xlabel('L')
pyplot.ylabel('Precompute Memory Cost')
pyplot.show()
```

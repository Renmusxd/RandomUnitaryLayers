```python
import numpy
import py_entropy
from matplotlib import pyplot
import scipy.special
```

# Nonphysical Chain

Here we pick $d^1_0 = d^1_1 = 2$ so that each $d^2_n > k$

```python
sizes = []
ls = numpy.arange(2,13)
for l in ls:
    g = py_entropy.GenericMultiDefectState(l, l//2, [2,2], parallel_matrix_mul=True, num_experiments=16)
    sizes.append(g.get_states().shape[0])
```

```python
theory = [2**l * scipy.special.comb(l, l//2) for l in ls]
```

```python
pyplot.plot(ls, sizes, 'x-')
pyplot.plot(ls, theory, '--')
pyplot.yscale('log')
pyplot.grid()
pyplot.xlabel('L')
pyplot.ylabel(r'$D_{N=0}$')
pyplot.title('Hilbert space dimension for N=L/2 sector')
pyplot.show()
```
